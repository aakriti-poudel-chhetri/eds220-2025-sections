```python
# Load libraries
import os
import pandas as pd

import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
# Import arctic communities data
fp = os.path.join('data','arctic_communities.geojson')
```

```python
arctic = gpd.read_file(fp)
```

```python
arctic.columns = arctic.columns.str.lower()
```

```python
arctic.head(10)
```

```python
arctic.geom_type.unique()
```

```python
arctic = arctic.set_index('admin')
```

```python
# if-else statement

if arctic.geom_type.unique().all() != 1:
    print(f"Multiple feature types {arctic.geom_type.unique()}")
else:
    print(f"All features are {arctic.geom_type.unique()}")
```

```python
def check_polygons(arctic):
    if arctic.geom_type.unique().all() != 1:
        print(f"Multiple feature types {arctic.geom_type.unique()}")
    else:
        print(f"All features are {arctic.geom_type.unique()}")
```

```python
check_polygons(arctic)
```

```python
arctic = arctic.explode(column = 'geometry', index_parts = False).reset_index()
```

```python
check_polygons(arctic)
```
